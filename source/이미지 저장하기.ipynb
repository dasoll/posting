```python
import zipfile
import os
import shutil
```

```python
def createDirectory(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print("Error: Failed to create the directory.")
```

```python
path = '/Users/dasolseo/Downloads/picture'
```

```python
for dirpath, dirname, filenames in os.walk(path):
    for file in filenames:
        name, extension = os.path.splitext(file)
        
        print(name, extension)
        if extension == '.zip':
            createDirectory(os.path.join(dirpath, name))
            fantasy_zip = zipfile.ZipFile(os.path.join(dirpath, file))
            fantasy_zip.extractall(os.path.join(dirpath, name))
 
            fantasy_zip.close()
            os.remove(os.path.join(dirpath, file))
```

```python
extensions = []
createDirectory(os.path.join(path, 'all'))


for dirpath, dirname, filenames in os.walk(path):
    for file in filenames:
        name, extension = os.path.splitext(file)        
        if extension not in ['.json', '.html']:
            shutil.move(os.path.join(dirpath, file), os.path.join(os.path.join(path, 'all'), file) )
#         extensions.append(extension)
```
